# Text Emotion Analysis\n
\n
This notebook trains a machine learning model using `emotion_dataset_10000.csv` and saves the trained model as `mlmodel/text_emotion_model.pkl`.

```python
from pathlib import Path\n
import pickle\n
\n
import pandas as pd\n
from sklearn.feature_extraction.text import TfidfVectorizer\n
from sklearn.linear_model import LogisticRegression\n
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix\n
from sklearn.model_selection import train_test_split\n
from sklearn.pipeline import Pipeline
```

```python
dataset_path = Path('emotion_dataset_10000.csv')\n
model_path = Path('mlmodel/text_emotion_model.pkl')\n
\n
df = pd.read_csv(dataset_path)\n
df = df.dropna(subset=['text', 'emotion'])\n
df.head()
```

```python
df['emotion'].value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(\n
    df['text'],\n
    df['emotion'],\n
    test_size=0.2,\n
    random_state=42,\n
    stratify=df['emotion'],\n
)\n
\n
model = Pipeline([\n
    ('tfidf', TfidfVectorizer(ngram_range=(1, 2), min_df=2)),\n
    ('classifier', LogisticRegression(max_iter=1000, class_weight='balanced')),\n
])\n
\n
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)\n
print('Accuracy:', accuracy_score(y_test, predictions))\n
print(classification_report(y_test, predictions))\n
confusion_matrix(y_test, predictions)
```

```python
with model_path.open('wb') as model_file:\n
    pickle.dump(model, model_file)\n
\n
model_path
```
